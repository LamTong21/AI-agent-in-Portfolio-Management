### Preparation

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller
from statsmodels.stats.diagnostic import acorr_ljungbox
from arch import arch_model
# Yêu cầu cài đặt thư viện hurst: pip install hurst
from hurst import compute_Hc
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel

def get_log_returns(prices_df):
    """Tính toán lợi suất logarit."""
    # Thay thế các giá trị 0.0 bằng NaN trước khi tính log để tránh log(0)
    prices_df = prices_df.replace(0.0, np.nan)
    log_returns = np.log(prices_df / prices_df.shift(1))
    return log_returns.dropna(how='all')

returns_df = get_log_returns(pricing_data_board)

returns_df.head()
```

## Step 2. Feature Engineering

```python
# 2.2. Ljung-Box (Bảo vệ Zero-Variance)
def ljung_box_pvalue(x):
    try:
        x_clean = x.dropna()
        # Nếu không đủ dữ liệu hoặc chuỗi nằm ngang (phương sai = 0), bỏ qua
        if len(x_clean) < window or x_clean.std() == 0:
            return np.nan
        res = acorr_ljungbox(x_clean, lags=[5], return_df=True)
        return res['lb_pvalue'].iloc[0]
    except:
        return np.nan

# 2.3. Fractal Dynamics (Sửa kind='change' và tăng window)
def calculate_hurst(series):
    clean_series = series.dropna()
    if len(clean_series) < 50:
        return np.nan
    try:
        # Lưu ý quan trọng: Dữ liệu truyền vào đang là lợi suất (returns),
        # nên kind phải là 'change' thay vì 'price' để thư viện tự cộng dồn (cumsum)
        H, _, _ = compute_Hc(clean_series, kind='change', simplified=True)
        return H
    except:
        return np.nan

def extract_quantitative_features(symbol_returns, window=20):
    features = pd.DataFrame(index=symbol_returns.index)

    # 2.1. Rolling Moments
    features['Rolling_Skew'] = symbol_returns.rolling(window=window).skew()
    features['Rolling_Kurt'] = symbol_returns.rolling(window=window).kurt()

    features['LB_PValue'] = symbol_returns.rolling(window=window).apply(ljung_box_pvalue, raw=False)

    # Tăng window lên 100 để đảm bảo Hurst chia scale ổn định
    hurst_window = 100
    features['Hurst_Exp'] = symbol_returns.rolling(window=hurst_window).apply(calculate_hurst, raw=False)

    # 2.4. Động lượng & Biến động
    features['Volatility'] = symbol_returns.rolling(window=window).std()
    features['Momentum_1D'] = symbol_returns.shift(1)

    return features
```

## Step 3. GARCH(1,1) -- Persistence

```python
def assess_garch_persistence(symbol_returns):
    """
    Fit mô hình GARCH(1,1) và trả về hệ số bám dính (persistence = alpha + beta).
    Chỉ chạy trên tập training hoặc cửa sổ tĩnh để tránh look-ahead bias.
    """
    try:
        # Scale dữ liệu lên 100 lần để optimizer của thuật toán hội tụ tốt hơn
        am = arch_model(symbol_returns.dropna() * 100, vol='Garch', p=1, q=1, rescale=False)
        res = am.fit(disp='off')

        alpha = res.params.get('alpha[1]', 0)
        beta = res.params.get('beta[1]', 0)
        persistence = alpha + beta
        return persistence
    except:
        return np.nan
```

## Step 4. Dimensionality & Selection

```python
def pipeline_for_symbol(symbol, returns_df):
    """Chạy toàn bộ quy trình có kiểm soát ngoại lệ."""
    ret_series = returns_df[symbol].dropna()

    if len(ret_series) < 150: # Bỏ qua các mã mới niêm yết, không đủ dữ liệu train
        return None

    # Bước 1: Stationarity Testing
    adf_stat, adf_pvalue, _, _, _, _ = adfuller(ret_series)

    # Bước 2: Feature Engineering
    features = extract_quantitative_features(ret_series)
    features['Target_Ret'] = ret_series.shift(-1)

    # CẢI TIẾN QUAN TRỌNG:
    # 1. Bỏ đi các feature (cột) bị hỏng hoàn toàn (ví dụ Hurst lỗi toàn tập)
    features = features.dropna(axis=1, how='all')
    # 2. Lúc này mới drop các dòng có chứa NaN (từ rolling window sinh ra)
    features = features.dropna(axis=0, how='any')

    if features.empty:
        return None

    X = features.drop(columns=['Target_Ret'])
    y = features['Target_Ret']

    # Bước 3: Dimensionality & Selection
    rf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    rf.fit(X, y)

    selector = SelectFromModel(rf, prefit=True)
    selected_features = X.columns[(selector.get_support())]

    importance = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)

    return {
        'ADF_pvalue': adf_pvalue,
        'Selected_Features': list(selected_features),
        'Feature_Importance': importance
    }
```

## Finaly Processing

```python
validation_dataset = pd.DataFrame()

# Ví dụ: Đánh giá persistence cho mã PNJ
for sym in hose_tickers:
    if sym in returns_df.columns:
        pnj_returns = returns_df[sym].dropna()
        pnj_persistence = assess_garch_persistence(pnj_returns)
        validation_dataset.loc["persistence", sym] = pnj_persistence
    else:
        print(f"Mã {sym} bị loại do không đủ dữ liệu hoặc chuỗi tĩnh.")


results = {}
# Gọi lại quy trình kiểm tra
for sym in hose_tickers:
    if sym in returns_df.columns:
        res = pipeline_for_symbol(sym, returns_df)
        if res is not None:
            results[sym] = res

            for key, value in res.items():
                if key != "Feature_Importance":
                    if isinstance(value, np.ndarray):
                        validation_dataset.loc[key, sym] = value
                    else:
                        validation_dataset.loc[key, sym] = str(value)

        else:
            print(f"Mã {sym} bị loại do không đủ dữ liệu hoặc chuỗi tĩnh.")

"""
if 'HSG' in results:
    print("\nKết quả lựa chọn đặc trưng cho HSG:")
    print(results['HSG']['Feature_Importance'])
else:
    print("\nKhông thể trích xuất đặc trưng cho HSG. Hãy kiểm tra lại phân phối dữ liệu đầu vào của mã này.")
"""
```

## Grouping Process

```python
import pandas as pd
import ast

def build_regime_clustering_pipeline(validation_dataset_df):
    """
    Thực thi 4 bước meta-clustering từ bảng kết quả validation.
    Giả định: validation_dataset_df là DataFrame gốc với các cột là mã cổ phiếu
    và các dòng là 'persistence', 'ADF_pvalue', 'Selected_Features'.
    """
    # Chuyển vị (Transpose) để mã cổ phiếu thành Index, các features thành cột
    df_transposed = validation_dataset_df.T

    # Chỉ giữ lại cột cần thiết và loại bỏ các mã bị lỗi (None/NaN)
    df_features = df_transposed[['Selected_Features']].dropna().copy()

    # ==========================================
    # Bước 1: Tiền xử lý và Tính toán tần suất
    # ==========================================
    # Đếm số lượng mã cổ phiếu cho mỗi tổ hợp đặc trưng
    freq_map = df_features['Selected_Features'].value_counts().to_dict()

    # Ánh xạ (Map) giá trị tần suất (N) ngược lại cho từng mã cổ phiếu
    df_features['Count_N'] = df_features['Selected_Features'].map(freq_map)

    # ==========================================
    # Bước 2 & 3: Thiết lập màng lọc và Phân loại
    # ==========================================
    def classify_regime(row):
        count_n = row['Count_N']
        feature_str = row['Selected_Features']

        # Bước 2: Lọc nhóm nhiễu (Threshold N < 5)
        if count_n < 5:
            return 'Noise_Discard'

        # Xử lý chuỗi an toàn: Chuyển "['Volatility']" thành cấu trúc set thực sự
        try:
            # Dùng ast.literal_eval an toàn hơn eval() rất nhiều
            feature_set = set(ast.literal_eval(feature_str))
        except (ValueError, SyntaxError):
            return 'Parsing_Error'

        # Bước 3: Phân loại theo cơ chế tập hợp (Ưu tiên từ cao xuống thấp)
        tail_risk_features = {'Rolling_Skew', 'Rolling_Kurt'}

        if len(feature_set.intersection(tail_risk_features)) > 0:
            return 'Higher_Moments'
        elif 'Hurst_Exp' in feature_set:
            return 'Fractal_Trending'
        elif feature_set.issubset({'Volatility', 'Momentum_1D'}):
            return 'Base_Core'
        else:
            return 'Unclassified'

    # Áp dụng hàm phân loại cho toàn bộ dataframe
    df_features['Market_Regime'] = df_features.apply(classify_regime, axis=1)

    # ==========================================
    # Bước 4: Tích hợp vào Pipeline (Tạo Dictionary)
    # ==========================================
    # Trích xuất ra một dictionary: Key là Mã cổ phiếu, Value là Nhóm mô hình
    regime_router = df_features['Market_Regime'].to_dict()

    # Thống kê nhanh kết quả phân cụm
    summary_stats = df_features['Market_Regime'].value_counts()

    return regime_router, summary_stats, df_features

regime_router, stats, detailed_df = build_regime_clustering_pipeline(df)
print(stats)
```
